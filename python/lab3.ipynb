```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_results(filename: str) -> np.ndarray:
    return np.loadtxt("lab3_results/" + filename)
```

```python
def add_results_to_plot(
    ax: plt.Axes, name: str, times: np.ndarray[float], thread_counts: np.ndarray[int]
) -> None:
    ax.scatter(thread_counts, times, label=name)
    ax.set_xlabel("Thread Count")
    ax.set_ylabel("Time [s]")
    ax.legend()
```

```python
ax: plt.Axes = plt.subplot()
ax.set_title("Results for variants: 2, 3, 5, 6")

thread_count: np.ndarray[int] = np.arange(5, 101, 5)

results2: np.ndarray[float] = load_results("results2.txt")
add_results_to_plot(ax, "variant 2", results2, thread_count)

results3: np.ndarray[float] = load_results("results3.txt")
add_results_to_plot(ax, "variant 3", results3, thread_count)

results5: np.ndarray[float] = load_results("results5.txt")
add_results_to_plot(ax, "variant 5", results5, thread_count)

results6: np.ndarray[float] = load_results("results6.txt")
add_results_to_plot(ax, "variant 6", results6, thread_count)
```
